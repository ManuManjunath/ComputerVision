```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# To create an empty bkack image
blank_img = np.zeros(shape= (512, 512, 3), dtype = np.int16)

plt.imshow(blank_img)
```

```python
# Drawing shapes on the above image.
# Rectangle
cv2.rectangle(blank_img, pt1 = (350, 50), pt2 = (450, 150), color = (255, 0, 0), thickness = 2)

# Circle
cv2.circle(blank_img, center = (100, 100), radius = 25, color = (255, 0, 0), thickness = 5)
#Circle with colour filled --> Set thickness = -1
cv2.circle(blank_img, center = (400, 100), radius = 25, color = (0, 0, 255), thickness = -1)

# Line
cv2.line(blank_img, pt1 = (0, 0), pt2 = (100, 100), color = (0, 255, 0), thickness = 2)

# Adding text
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, 
            text = "Hello World", 
            org = (0, 512), # Bottom left corner of the text in the image
            fontFace = font, 
            fontScale = 2, 
            color = (255, 255, 255), 
            thickness = 3,
           lineType = cv2.LINE_AA)

plt.imshow(blank_img)
```

```python
# Drawing custom polygons
blank_img = np.zeros(shape = (512, 512, 3), dtype = np.int32)

vertices = np.array([[100, 100], [400, 150], [400, 400], [100, 400]], dtype = np.int32)
#vertices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype = np.int32)
# This is a 2 dimensional array.
print(vertices.shape)
print(vertices)

# However, opencv would require this to be 3 dimensional.
points = vertices.reshape(-1, 1, 2)

# Now this is 3 dimensional
print("After reshaping..")
print(points.shape)
print(points)
```

```python
cv2.polylines(blank_img, [points], isClosed = True, color = (225, 0, 0), thickness = 2)
plt.imshow(blank_img)
```
