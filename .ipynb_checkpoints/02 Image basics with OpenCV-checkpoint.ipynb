```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image

img_file = "data/00-puppy.jpg"

%matplotlib inline
```

```python
img = cv2.imread(img_file)
print(type(img))
print(img.shape)
plt.imshow(img)
# Matplotlib colour channels --> Red, Blue, Green
# OpenCV colour channels --> Blue, Green, Red
```

```python
# To fix the colouring
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
# To view in gray
img_gray = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)
print(img_gray.shape)
plt.imshow(img_gray, cmap = "gray")
```

```python
# Resizing images
resized = cv2.resize(fix_img, (600, 200))

# View the shape
print(resized.shape)
# The shape requested was (600, 200) But we get (200, 600). Because X&Y axes in OpenCV and Numpy are reversed!

plt.imshow(resized)
```

```python
# Resizing by ratio
width_ratio = 0.5
height_ratio = 1.5
resized_ratio = cv2.resize(fix_img, (0, 0), fix_img, width_ratio, height_ratio)

print(resized_ratio.shape)
# Cross check the new size

plt.imshow(resized_ratio)
```

```python
# To flip an image along the horizontal axis
img_flipped = cv2.flip(fix_img, 0)
# 0 flips it across the horizontal axis. So shows the image upside down
plt.imshow(img_flipped)
```

```python
# To flip an image along the vertical axis
img_mirrored = cv2.flip(fix_img, 1)
# 1 flips it across the vertical axis. So shows the mirrored image
plt.imshow(img_mirrored)
```

```python
# To flip an image along the horizontal and vertical axis
img_mirrored_flip = cv2.flip(fix_img, -1)
# -1 flips it across the horizontal and vertical axes. So shows the mirrored image upside down
plt.imshow(img_mirrored_flip)
```

```python
# To adjust the image viewing canvas size
fig = plt.figure(figsize = (10, 8))
ax = fig.add_subplot(111)
ax.imshow(fix_img)
```

```python
# To save an image.
cv2.imwrite("data/flipped.jpg", img_mirrored_flip)
```
