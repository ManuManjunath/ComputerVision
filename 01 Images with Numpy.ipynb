```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

%matplotlib inline
```

```python
# Reading an image
pic = Image.open("DATA/00-puppy.jpg")
#pic
print(type(pic))

# Converting pic to an array
pic_arr = np.asarray(pic)
print(type(pic_arr))

print(pic_arr.shape)
# 3 dimensional array --> 1300 px height, 1950 px width, 3 colour channels (R, G, B)
```

```python
# To view the image
plt.imshow(pic_arr)
```

```python
# To only view the red channel
pic_red = pic_arr[:,:,0]

# View the Red channel array
print(pic_red)

plt.imshow(pic_red)
```

```python
# To view the green channel
pic_green = pic_arr[:,:,1]

# View the Green channel array
print(pic_green)

plt.imshow(pic_green)
```

```python
# To view the blue channel
pic_blue = pic_arr[:,:,2]
plt.imshow(pic_blue)
```

```python
# Make the image blue (Remove Red and Green channels)
pic_copy = pic_arr.copy()
# R G B
pic_copy[:,:,0] = 0
pic_copy[:,:,1] = 0
plt.imshow(pic_copy)
```

```python
# Make the image red (Remove Blue and Green channels)
pic_copy = pic_arr.copy()
# R G B
pic_copy[:,:,1] = 0
pic_copy[:,:,2] = 0
plt.imshow(pic_copy)
```

```python
# Make the image green (Remove Blue and Red channels)
pic_copy = pic_arr.copy()
# R G B
pic_copy[:,:,0] = 0
pic_copy[:,:,2] = 0
plt.imshow(pic_copy)
```

```python
# To view a single channel image in gray
plt.imshow(pic_blue, cmap = 'gray')
```
